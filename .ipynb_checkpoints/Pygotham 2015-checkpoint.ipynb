# Formal Information System

## Information Sets

Information sets are simply named dictionaries.

```python
class InformationSet(dict):
    def __init__(self,name, *args):
        self.name = name
        dict.__init__(self,*args)
```

```python
USCF = InformationSet('USCF', {'tournaments': ['t1','t2','t3']})
```

As a rule information sets are always named in all caps with the name of the data source.  One data source, one information set.

```python
FIDE = InformationSet('FIDE', {'tournaments': ['t1','t2','t3']})
```

## Collections

An inference set begins as a collection of information sets (a dictionary of dictionaries).  These are the information sets from which the inference will come.

```python
class Collection(dict):
    def __init__(self, *args):
        dict.__init__(self, {h.name: h for h in args})
        self.original = self.keys()
```

```python
collection = Collection(USCF, FIDE)
```

```python
collection
```

```python
collection['FIDE']['tournaments']
```

A collection class could have some other helpful behavior.  A dictionary of dictionaries makes it hard to access the underlying data.  Consider adding two helpful functions:
- Replace generic 'get' functionality with one optimized for dict of dict searches.
- Add a flatten function that flattens a dict of dicts into a dict with hyphens added to the name.

```python
class Collection(dict):
    def __init__(self, *args):
        dict.__init__(self, {h.name: h for h in args})
        self.original = self.keys()
    
    def __getitem__(self, key):
        if '-' in key:
            source, name = key.split('-')
            return super(Collection, self).__getitem__(source).get(name, None)
        else:
            return super(Collection, self).get(key, None)
    
    def flatten(self):
        flatten_keys = [h for h in self.keys() if h not in self.original]
        flattened_dict = dict([(h, self[h]) for h in flatten_keys])
        #add origin files
        for infoSet in self.original:
            for key, value in self[infoSet].items():
                new_key = "%s-%s"%(infoSet, key)
                flattened_dict.update({new_key:value})
        return flattened_dict
```

## InferenceSet

An inference set is a named collection.

```python
class InferenceSet(Collection):
    def __init__(self, name, *args):
        Collection.__init__(self, *args)
        self.name = name
```

```python
NumTournaments = InferenceSet('NumTournaments', USCF, FIDE)
```

As a rule inference sets are named with the first letter of each word capitalized.

```python
NumTournaments
```

Inference sets make it possible to reach higher levels of complexity by inferring something new from an underlying set.  This inferring functionality is coded separately as process functions.  An inference set starts as a collection of original information sets, processes them by calling a process function, and is then a collection of both the original data sources and the new inferred data.

## Process functions

Process functions operate on an InferenceSet and return a dictionary.

```python
def count_tournaments(InferSet):
    uscf_tournaments = InferSet['USCF-tournaments']
    fide_tournaments = InferSet['FIDE-tournaments']
    num_tournaments = len(uscf_tournaments)+len(fide_tournaments)
    return {'num_tournaments': num_tournaments}
```

```python
count_tournaments(NumTournaments)
```

```python
NumTournaments
```

```python
NumTournaments.update(count_tournaments(NumTournaments))
```

```python
NumTournaments
```

This can be formalized by writing a process function within the InferenceSet class.

```python
class InferenceSet(Collection):
    def __init__(self, name, *args):
        Collection.__init__(self, *args)
        self.name = name
    
    def process(self, func, input=''):
        if self.__repr__()!='{}':
            if input:
                self.update(func(input))
            else:
                self.update(func(self))
```

InferenceSet includes both new inferred variables and the original data.  When writing to a database it could cause duplicated entries.

- Add an 'inferred' function that returns only the new inferred variables.

```python
class InferenceSet(Collection):
    def __init__(self, name, *args):
        Collection.__init__(self, *args)
        self.name = name
    
    def inferred(self, return_informationset = True, return_dict = False):
        inferred_keys = [h for h in self.keys() if h not in self.original]
        inferred_dict = dict([(h, self[h]) for h in inferred_keys])
        if return_informationset:
            return InformationSet(self.name, inferred_dict)
        elif return_dict:
            return inferred_dict
        else:
            return inferred_keys
    
    def process(self, func, input=''):
        if self.__repr__()!='{}':
            if input:
               self.update(func(input))
            else:
                self.update(func(self))
```

```python
NumTournaments = InferenceSet('NumTournaments', USCF, FIDE)
```

```python
NumTournaments.process(count_tournaments)
```

```python
NumTournaments.inferred()
```

# PyCount

Like 'Plycount' but written in Python!

For a given user's USCF rating ID:
- collect all user's tournaments from the USCF website.
- save in an information set.

Then calculate:
- number of tournaments.
- a user's best tournament.

Then combine with a Flask front end:
- display information in an easy-to-use and helpful way.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

id = raw_input('Enter USCF id: ')


count = 0
data = {}
for i in range(4):
    h = requests.get('http://www.uschess.org/msa/MbrDtlTnmtHst.php?'+id+'.'+str(i))
    soup = BeautifulSoup(h.text)

    for h in soup.find_all('tr'):
        if h.get('bgcolor',0)=='FFFF80' or h.get('bgcolor',0)=='FFFFC0':
            events = h.find_all('td')
            for event in events:
                if count%5==0 and count !=0:
                    id = id
                    date = events[0].contents[0]
                    info = events[1].contents[0].text
                    data[count] = {'id':id, 'date': date, 'info':info}
                count +=1

df = pd.DataFrame.from_dict(data, orient='index')
df = df.drop_duplicates('date')
```

```python
USCF = InformationSet('USCF', {'tournaments':df.to_dict('records')})
```

```python
NumTournaments = InferenceSet('NumTournaments', USCF)
```

```python
NumTournaments.process(count_tournaments)
```
